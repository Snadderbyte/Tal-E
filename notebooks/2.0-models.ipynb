# Creating some simple models using the interim dataset.

```python
from fastai.vision.all import *
import numpy
```

```python
path = Path('../data/final')
```

```python
files = get_files(path, recurse=True, folders=None)
```

```python
# Checking if all files are loaded
len(files)
```

## Get the labels

```python
import re
pattern = r'\d'

def get_label(file):
    return file[re.search(pattern, file).start()]
```

```python
get_label(f'{files[0]}')
```

```python
dls = ImageDataLoaders.from_name_func(path, files, get_label)
```

```python
dls.show_batch()
```

```python
learn = vision_learner(dls, arch=resnet18, metrics=accuracy)
```

```python
learn.fine_tune(2)
```

```python
learn.show_results()
```
